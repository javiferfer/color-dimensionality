```python
import colorsys
import glob
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import pandas as pd
import seaborn as sns
```

```python
img_s = sorted(glob.glob('../colors/output/img/*.png'))
```

```python
colors_rgb = []
for filename in img_s:
    img = np.array(Image.open(filename)) / 256.0
    colors_rgb.append(img[200, 150, :3])
colors_rgb = np.array(colors_rgb)
print(colors_rgb)
```

#### HSV representation

```python
hsv_s = []
for i in range(len(colors_rgb)):
    r = colors_rgb[i, 0]
    g = colors_rgb[i, 1]
    b = colors_rgb[i, 2]
    hsv = colorsys.rgb_to_hsv(r, g, b)
    hsv_s.append(hsv)
hsv_s = np.array(hsv_s)

sns.distplot(hsv_s[:, 0], kde=False, bins=20)
print(hsv_s[:, 0])
```

```python
plt.scatter(hsv_s[:, 0], hsv_s[:, 1], c=colors_rgb)
plt.show()
plt.scatter(hsv_s[:, 0], hsv_s[:, 2], c=colors_rgb)
plt.show()
```

```python
plt.style.use('default')
fig = plt.figure(figsize=(6, 4), tight_layout=True, dpi=100)
ax = fig.add_subplot(projection='3d')
ax.scatter(hsv_s[:, 0], hsv_s[:, 1], hsv_s[:, 2], c=colors_rgb, s=25)
ax.set_xlabel('Hue')
ax.set_ylabel('Saturation')
ax.set_zlabel('Value')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.grid(False)
plt.savefig('output/fig/colors_hsv.png')
```

#### Displaying images based on the hue

```python
for i in range(0, 100, 1):
    (r, g, b) = colorsys.hsv_to_rgb(i/100, 1.0, 1.0)
    print(i/100)
    color = [r, g, b]
    color = [[color]*10]*10
    print(np.shape(color))
    plt.figure(figsize=(2, 2))
    plt.imshow(color)
    plt.show()
```
